```python
import os
import pandas as pd
import numpy as np
import math
import datetime as dt

# For Evalution we will use these library

from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score, r2_score
from sklearn.metrics import mean_poisson_deviance, mean_gamma_deviance, accuracy_score
from sklearn.preprocessing import MinMaxScaler

# For model building we will use these library

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.layers import LSTM
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau



# For PLotting we will use these library

import matplotlib.pyplot as plt
from itertools import cycle
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
%matplotlib inline
from plotly.offline import init_notebook_mode,iplot
init_notebook_mode(connected=True)
```

```python
df=pd.read_csv("/content/ds.csv")
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values(by='Date', ascending=True, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
import numpy as np



# Identify numerical columns (excluding 'Date')
NumCols = df.columns.difference(['Date'])

# Step 1: Replace '#NUM!' with NaN
df[NumCols] = df[NumCols].replace('#NUM!', np.nan)

# Step 2: Remove non-numeric characters (keep digits, commas, periods, hyphens)
df[NumCols] = df[NumCols].replace({
    r'[^\d.,-]': '',  # Regex to keep only numbers and separators
}, regex=True)

# Step 3: Replace empty strings with NaN
df[NumCols] = df[NumCols].replace({'': np.nan})

# Step 4: Convert to float
df[NumCols] = df[NumCols].astype('float64')

# Step 5: Fill NaNs (if any)
df[NumCols] = df[NumCols].fillna(0)

# Verify the result
print("Processed data:")
print(df.head())
print(df.info())
```

```python
# Select all features (including price, interest rate, USD rates, etc.)
features = ['price', 'interest rate', 'usd_buy', 'usd_sell', 'Inflation Rate']

# Scale ALL features
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[features])

# Create sequences with MULTIVARIATE data (all features)
def create_sequences(data, lookback):
    X, y = [], []
    for i in range(lookback, len(data)):
        X.append(data[i-lookback:i, :])  # All features as input
        y.append(data[i, 0])            # 'price' as target
    return np.array(X), np.array(y)

X_train, y_train = create_sequences(scaled_data, lookback=30)
X_test, y_test = create_sequences(scaled_data, lookback=30)

print("X_train shape:", X_train.shape)  # Should be (n_samples, 30, 5)
print("y_train shape:", y_train.shape)  # (n_samples,)
```

```python
# ✅ Correct: Split before scaling
train_size = int(len(df) * 0.7)
train, test = df[0:train_size], df[train_size:]

# Scale train and test separately
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train[features])
test_scaled = scaler.transform(test[features])

# Create sequences
X_train, y_train = create_sequences(train_scaled, lookback=30)
X_test, y_test = create_sequences(test_scaled, lookback=30)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout, BatchNormalization
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping
from keras import regularizers

# Define the model
model = Sequential()

# Add the first LSTM layer with regularization and dropout
model.add(LSTM(64, activation='tanh', return_sequences=False,
               input_shape=(X_train.shape[1], X_train.shape[2]),
               kernel_regularizer=regularizers.l2(0.001)))  # L2 regularization
model.add(BatchNormalization())
model.add(Dropout(0.3))  # Increased dropout rate

# Add Dense layer after LSTM with regularization
model.add(Dense(32, activation='relu',
                kernel_regularizer=regularizers.l2(0.001)))  # L2 regularization
model.add(BatchNormalization())

# Add output layer
model.add(Dense(1))

# Compile with a lower learning rate
optimizer = Adam(learning_rate=0.00005, clipvalue=1.0)
model.compile(optimizer=optimizer, loss="mean_squared_error")

# Early stopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Summary of the model
model.summary()

# Fit the model with early stopping
history= model.fit(X_train, y_train, epochs=100, batch_size=32,
          validation_data=(X_test, y_test),
          callbacks=[early_stopping])
```

```python
import matplotlib.pyplot as plt

# Plot training & validation loss values
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Assuming you've already split your dataset and have X_test, y_test

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate using common metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calculate MSE, MAE, and R2
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print out the metrics
print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R²): {r2}")

# Optionally, visualize the results (Actual vs Predicted)
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(y_test, label="Actual Values", color='blue')
plt.plot(y_pred, label="Predicted Values", color='red')
plt.title("Actual vs Predicted Values")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.show()

```
